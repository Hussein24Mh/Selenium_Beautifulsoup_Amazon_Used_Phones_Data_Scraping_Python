```python
import csv
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

```python
def gen_url(term, page):
    '''Generate a url from passed term'''
    template = 'https://www.amazon.com/s?k={}&page={}'
    term = term.replace(' ','+')
    return template.format(term, page)
```

```python
def get_page_source(url):
    '''get page source from the passed url'''
    driver.get(url)
    return driver.page_source
```

```python
def get_phones_pages_links(term , pages):    
    for i in range(1,pages+1):
        url = gen_url(term, i)
        soup = BeautifulSoup(get_page_source(url), 'html.parser')
        current_products_list = soup.find_all('div', {'data-component-type':'s-search-result'})
        
        for j in range(0,len(current_products_list)):
            item = current_products_list[j]
            item_url = 'https://www.amazon.com/' + item.h2.a.get('href')
            all_details.append([item_url])
```
