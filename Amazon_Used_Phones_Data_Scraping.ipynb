```python
import csv
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

```python
def gen_url(term, page):
    '''Generate a url from passed term'''
    template = 'https://www.amazon.com/s?k={}&page={}'
    term = term.replace(' ','+')
    return template.format(term, page)
```

```python
def get_page_source(url):
    '''get page source from the passed url'''
    driver.get(url)
    return driver.page_source
```

```python
def get_phones_pages_links(term , pages):    
    for i in range(1,pages+1):
        url = gen_url(term, i)
        soup = BeautifulSoup(get_page_source(url), 'html.parser')
        current_products_list = soup.find_all('div', {'data-component-type':'s-search-result'})
        
        for j in range(0,len(current_products_list)):
            item = current_products_list[j]
            item_url = 'https://www.amazon.com/' + item.h2.a.get('href')
            all_details.append([item_url])
```

```python
def get_all_details():
    '''Now we have the likes to web page of each phone we get from get_phones_pages_links
       So here the will get all details for each one
    '''
    for i in range(len(all_details)):
        print(i)
        soup = BeautifulSoup(get_page_source(all_details[i][0]), 'html.parser')

        # this centercol have all the information we want to get
        centercol = soup.find('div', id="centerCol")
               
        # full tag of product store information
        try:
            storelinktag = centercol.find('a', id='bylineInfo')
            try:
                product_store_name = storelinktag.text[10:].strip()
            except:
                product_store_name = None

            try:
                product_store_link = 'https://www.amazon.com' + storelinktag.get('href').strip()
            except:
                product_store_link = None
        except:
            product_store_name = None
            product_store_link = None
            
        # this in another quick access div have many information we will extract
        try:
            qtable = centercol.find('table', 'a-normal a-spacing-micro')
        except:
            pass


        try:
            product_des = centercol.find('span', id='productTitle').text.strip()
        except:
            product_des = None
        
        try:
            product_rating = centercol.find('span', 'a-icon-alt').text[:3]
        except:
            product_rating = None
            
        try:
            product_price = centercol.find('span', {'data-a-color':'price'}).span.text.strip()
        except:
            product_price = None
        
        try:
            product_modlename = qtable.find('tr', "a-spacing-small po-model_name").find('td', 'a-span9').span.text.strip()
        except:
            product_modlename = None
        
        try:
            product_Carrier = qtable.find('tr', "a-spacing-small po-wireless_provider").find('td', 'a-span9').span.text.strip()
        except:
            product_Carrier = None
            
        try:
            product_Brand = qtable.find('tr', "a-spacing-small po-brand").find('td', 'a-span9').span.text.strip()
        except:
            product_Brand = None
        
        try:
            product_Form_Factor = qtable.find('tr', "a-spacing-small po-form_factor").find('td', 'a-span9').span.text.strip()
        except:
            product_Form_Factor = None
        
        try:
            product_Memory = qtable.find('tr', "a-spacing-small po-memory_storage_capacity").find('td', 'a-span9').span.text.strip()
        except:
            product_Memory = None
            
        try:
            product_os = qtable.find('tr', "a-spacing-small po-operating_system").find('td', 'a-span9').span.text.strip()
        except:
            product_os = None
            
        try:
            product_color = qtable.find('tr', "a-spacing-small po-color").find('td', 'a-span9').span.text.strip()
        except:
            product_color = None
        
        try:
            product_Cellular = qtable.find('tr', "a-spacing-small po-cellular_technology").find('td', 'a-span9').span.text.strip()
        except:
            product_Cellular = None
        
        try:
            product_scsc = qtable.find('tr', "a-spacing-small po-sim_card_slot_count").find('td', 'a-span9').span.text.strip()
        except:
            product_scsc = None
        
        try:
            product_Year = qtable.find('tr', "a-spacing-small po-model_year").find('td', 'a-span9').span.text.strip()
        except:
            product_Year = None
        
        try:
            product_Components = qtable.find('tr', "a-spacing-small po-included_components").find('td', 'a-span9').span.text.strip()
        except:
            product_Components = None

        # we have got 15 feature from the product page
        # we need to store them in our all_details nestedlist
    
        all_details[i].extend([product_des, product_store_name, product_store_link, product_rating, product_price,
                               product_modlename, product_Carrier, product_Brand, product_Form_Factor, product_Memory,
                               product_os, product_color, product_Cellular, product_scsc, product_Year,
                               product_Components
                               ])
```

```python
all_details = []

# open the driver service
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

get_phones_pages_links("used phones", 20)
print(len(all_details))

get_all_details()

# close the driver service
driver.quit()
```

```python
# save to csv
with open('Collected_data.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['ProductLink',
                     'Description', 'SellerStore', 'StoreLink',
                     'Rating', 'Price', 'ModleName',
                     'Carrier',
                     'Brand', 'FormFactor', 'Memory',
                     'OperatingSystem', 'Color', 'Cellular',
                     'SimCount', 'Year', 'AdditionalComponents'])
    writer.writerows(all_details)
```
